```python
#11.	Write a Pandas program to append a column of your choice to your DataFrame.

#files[Referee] = "Sylver Mo"

#create a match type column containing the resuts of the match
import numpy as np

conditions = [
    (file['HomeTeamGoals'] > file['AwayTeamGoals']),  
    (file['HomeTeamGoals'] < file['AwayTeamGoals']),  
    (file['HomeTeamGoals'] == file['AwayTeamGoals'])  
]

choices = ['Home Win', 'Away Win', 'Draw']

# Apply conditions and choices to create the new column
file['MatchType'] = pd.Series(np.select(conditions, choices))

# Display the updated DataFrame
print(file.head())
```
